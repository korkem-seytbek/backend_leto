```
import pandas as pd
import kagglehub

# Датасетті жүктеу
path = kagglehub.dataset_download("yasserh/titanic-dataset")
print("Path to dataset files:", path)

# CSV оқу
df = pd.read_csv(path + "/Titanic-Dataset.csv")
df.head()

```

```
df.head()
```

```
df.info()
df.describe()
```

```
# Әйел жолаушылар
women = df[df['Sex'] == 'female']
women.head()

# 18 жасқа дейінгі балалар
kids = df[df['Age'] < 18]
kids.head()

# Бірінші класстағы жолаушылар
first_class = df[df['Pclass'] == 1]
first_class.head()

# Тірі қалған әйелдер
survived_women = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
survived_women.head()
```

```
# 30 жастан асқан ер адамдар
df.query("Age > 30 and Sex == 'male'").head()

# 2 немесе 3 класта отырғандар
df.query("Pclass in [2, 3]").head()

# 18 жасқа дейінгі және тірі қалғандар
df.query("Age < 18 and Survived == 1").head()

# 50-ден көп ақша төлегендер
df.query("Fare > 50").head()
```
